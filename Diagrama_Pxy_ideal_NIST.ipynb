```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import jtplot submodule from jupyterthemes
from jupyterthemes import jtplot

# currently installed theme will be used to
# set plot style if no arguments provided
jtplot.style(theme='onedork')
```

```python
AC1 = [4.01814,1203.835,-53.226]
AC2 = [4.08245,1346.382,-53.508]	
```

```python
# composición inicial zi
z = np.linspace(0,1,num=11)
```

```python
def Psat1(T):
  return 10 ** ( AC1[0] - AC1[1] / ( T + AC1[2] ) )

def Psat2(T):
  return 10 ** ( AC2[0] - AC2[1] / ( T + AC2[2] ) )   
```

```python
x = z
T = 115 +273.15 # K
```

```python
P = x * Psat1(T) + ( 1 - x ) * Psat2(T)
y = x * Psat1(T)  / P
```

```python
plt.plot(x,P,'w',y,P,'y')
```

```python
fig = plt.figure(dpi=100, figsize=(5,5), tight_layout=True)
plt.style.library['ggplot']
plt.plot(x,P,'w',y,P,'y')
plt.xlim(0,1)
plt.title('Diagrama P$_{xy}$ benceno - tolueno a 100 $^\circ$C')
plt.ylabel('Presión (Pa)')
plt.xlabel('$x_1$')
plt.grid('on')
plt.show()
```
